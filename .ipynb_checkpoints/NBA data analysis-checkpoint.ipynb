```python
def get_year(dt):
    return dt['Year']

data_by_year = (get_year(data) > 2000)
```

```python
by_year = data[data_by_year]
by_year
```

```python
player = by_year['Player']
steals = by_year['STL']
turnover = by_year['TOV']
assists = by_year['AST']
```

```python
plot(steals, '^', ms=1)
xlim(15000, 24000)
ylim(0, 250)
```

```python
plot(turnover, '.', ms=1)
xlim(15000, 24000)
ylim(0, 400)
```

```python
plot(assists, '.', ms=1)
xlim(15000, 24000)
ylim(0, 900)
```

```python
plot(assists, turnover, '.', ms=1)
ylabel('turnover')
xlabel('assists')
# xlim(15000, 24000)
# ylim(0, 250)
# plot(turnover, '.', ms=1)
# xlim(15000, 24000)
# ylim(0, 400)
```
